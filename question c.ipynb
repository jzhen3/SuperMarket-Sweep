### TSP Formulation

```python
from pulp import *
```

```python
prob = LpProblem("Supermarket Sweep", LpMaximize)
```

```python
# decision variables
x1 = [[0 for i in range(58)] for j in range(58)]
x2 = [[0 for i in range(58)] for j in range(58)]
y1 = [0 for i in range(58)]
y2 = [0 for i in range(58)]
t1 = [[0 for i in range(58)] for j in range(58)]
t2 = [[0 for i in range(58)] for j in range(58)]
c1 = [0 for i in range(58)]
c2 = [0 for i in range(58)]
```

```python
for i in range(58):
    for j in range(58):
        x1[i][j] = LpVariable(str('x1_'+str(i)+'_'+str(j)), 0, 1, LpInteger)
        x2[i][j] = LpVariable(str('x2_'+str(i)+'_'+str(j)), 0, 1, LpInteger)
for i in range(58):
    y1[i] = LpVariable(str('y1_'+str(i)), 0, None, LpContinuous)
    y2[i] = LpVariable(str('y2_'+str(i)), 0, None, LpContinuous)
for i in range(58):
    for j in range(58):
        t1[i][j] = LpVariable(str('t1_'+str(i)+'_'+str(j)), 0, None, LpContinuous)
        t2[i][j] = LpVariable(str('t2_'+str(i)+'_'+str(j)), 0, None, LpContinuous)
for i in range(58):
    c1[i] = LpVariable(str('c1_'+str(i)), 0, 1, LpInteger)
    c2[i] = LpVariable(str('c2_'+str(i)), 0, 1, LpInteger)
```

```python
# x1 = np.array(x1)
# x2 = np.array(x2)
# y1 = np.array(y1)
# y2 = np.array(y2)
# t1 = np.array(t1)
# t2 = np.array(t2)
# c1 = np.array(c1)
# c2 = np.array(c2)
```

```python
# (c1 + c2) * p
# c1 * p
```

```python
# objective function
obj = lpSum((c1[i] + c2[i]) * p[i] for i in range(58))
# obj = sum(c1 * p)
prob += obj
```

```python
# constraints

# shopper leaves item i
for i in range(57):
    prob += lpSum(x1[i][1:])==c1[i]
    prob += lpSum(x2[i][1:])==c2[i]

# shopper reaches item j
for j in range(1,58):
    prob += lpSum(x1[i][j] for i in range(57))==c1[j]
    prob += lpSum(x2[i][j] for i in range(57))==c2[j]

# both shoppers leave the start node
prob += c1[0]==1
prob += c2[0]==1

# both shoppers reach the end node
prob += c1[57]==1
prob += c2[57]==1

# do not return to the start node
for i in range(58):
    prob += x1[i][0]==0
    prob += x2[i][0]==0

# do not start from the end node
for j in range(58):
    prob += x1[57][j]==0
    prob += x2[57][j]==0

# do not move to oneself
for i in range(58):
    prob += x1[i][i]==0
    prob += x2[i][i]==0

# for each item, at most one shopper reaches it
for i in range(1,57):
    prob += c1[i]+c2[i]<=1

# each shopper can pick up at most 10 items
prob += lpSum(c1[1:57])<=10
prob += lpSum(c2[1:57])<=10

# if x_ij = 0, then t_ij = 0
for i in range(58):
    for j in range(58):
        prob += t1[i][j]<=150*x1[i][j]
        prob += t2[i][j]<=150*x2[i][j]

# boundary conditions
prob += y1[0]==0
prob += y2[0]==0

# y_j in terms of t_ij
for j in range(1,58):
    prob += y1[j]==lpSum(t1[i][j] for i in range(57))
    prob += y2[j]==lpSum(t2[i][j] for i in range(57))

# the time of picking the item that follows item j
for j in range(57):
    prob += lpSum(t1[j][1:])==y1[j]+lpSum(d[j][i]*x1[j][i] for i in range(1,58))
    prob += lpSum(t2[j][1:])==y2[j]+lpSum(d[j][i]*x2[j][i] for i in range(1,58))

# time constraint
    prob += y1[57]+2*lpSum(c1[1:57])<=60
    prob += y2[57]+2*lpSum(c2[1:57])<=60
```
